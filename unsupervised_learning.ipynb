### 1. 머신러닝의 비지도학습의 대표적인 모델과 기술

$𝒙^1, 𝒙^2, … , 𝒙^𝑁$의 샘플에서 hidden stucture을 찾는 방법을 비지도학습(Unsupervised Learning)이라고 한다. 지도학습과 달리 정답 label이 없는 데이터를 비슷한 특징끼리 군집화하여 새로운 데이터에 대한 결과를 예측한다. 대표적으로 clustering, demesionality reduction, association analysis 등에 활용된다.

#### Clustering
: 서로 유사한 속성을 갖는 데이터를 같은 군집으로 묶어주는 과정이다. 거대한 양의 데이터들을 유사한 속성으로 묶어 숨겨진 유의미한 구조를 파악할 수 있다는 장점이 있다. 지도학습의 분류(Classification)와 유사하지만, **정답 label 없이** 데이터를 비슷한 특징으로 군집화하여 새로운 데이터에 대한 결과를 예측한다는 점에서 다르다.

사전에 군집의 개수를 정하지 않고, 유사한 데이터를 차례로 묶어 계층적으로 결합하는 계층적 클러스터링(Hierarchical Clustering)과 사전에 군집의 개수를 정하는 분할적 클러스터링(Partitional)으로 구분할 수 있다. 또한, 계층적 클러스터링은 데이터가 특정 중심을 기준으로 분포한다고 가정하는 중심 기반(Center-based) 방법과 데이터가 서로 근접하게 분포한다고 가정하는 밀도 기반(Density-based) 방법으로 나눌 수 있다.

##### K-Means Clustering
- K개의 군집을 나누고, 각 그룹에서 랜덤으로 정한 중심(Means)으로부터 1. 가까운 데이터는 동일한 카테고리로 군집화된다. 이후, 2. 각 그룹의 중심은 군집 내 데이터들의 평균값으로 계산이 되어 이동한다. 1번과 1번 과정을 반복하여 중심의 이동이 없으면 종료한다.
- 속도가 비교적 빠르지만, 군집의 개수를 정하기 까다롭다는 단점이 있다.

##### DBSCAN

- K-Means Clustering의 단점을 보완하여 밀도를 고려한 알고리즘으로, 주로 이상치가 있는 대규모 데이터에 적용할 수 있다.
- Epsilon은 각각의 데이터의 반경인 원의 반지름을 나타내는 값으로 해당 반경 안에 속하는 데이터는 이웃으로 정의한다.
- Min points은 최소 군집의 크기를 나타내며, 각각의 데이터가 군집으로 정의되기 위해 필요한 최소한의 이웃 수다(자기 자신 포함).
- 각 데이터는 Core(중심점), Border(경계점), Noise(잡음, 이상치)으로 구분된다. Core은 epsilon 내에 min points 이상의 점이 존재하는 중심점을 말한다. Border는 군집의 중심점은 아니지만 군집에 속해있는 점을 의미한다. Noise는 군집에 포함되지 못하는 점이다.
- K-Means Clustering과 달리 군집의 개수를 사전에 지정하지 않고 자동으로 찾아준다는 점에서 유용하다. 또한, K-Means Clustering의 경우 이상치가 평균 계산에 사용되어 새로운 군집의 중심점이 왜곡될 수 있지만, DBSCAN은 이상치를 구분해서 제외한다는 점에서 의미 없는 군집이 생성되는 것을 방지할 수 있다. 그러나 많은 연산이 필요해서 처리 속도가 느리다는 단점이 있다.

##### Gaussian Mixture Model
- 데이터가 여러 다른 모양의 가우시안 분포(Gaussian Distribution)로 결합되어 있다는 가정 하에 개별 데이터를 동일한 가우시안 분포별로 묶어주는 확률 모델이다.
- 기하학적인 모양의 군집, 서로 겹치는 군집에 대해서도 클러스터링이 잘 수행된다는 점에서 K-Means Clustering의 단점을 보완한다.
- 개별 데이터에 대해 각 정규 분포(가우시안 분포)에 속할 확률을 계산하여 할당하고, 최대 우도 추정법으로 개별 정규 분포의 평균과 분산 추정을 반복하며 학습이 진행된다.
- 크게 예측(Expectation)과 최대화(Maximization) 2단계로 구분된다. 예측 단계에서는 개별 데이터 포인트가 각 정규 분포로부터 생성되었을 가능도(likelihood)를 계산하여, 가장 높은 확률을 가진 정규 분포에 할당한다. 개별 데이터 포인트를 모두 할당한 후, 각 그룹의 데이터 포인트를 이용하여 Maximum Likelihood Estimation(최대 우도 추정)으로 모분포의 모평균과 모분산을 추정한다. 개별 데이터들의 소속과 정규 분포의 모수(평균과 분산)가 변하지 않을 때까지 1,2 단계를 반복 수행한다.

#### Demensionality Reduction

##### PCA(Principle Component Analysis)
- 차원의 저주(Curse of Dimensionality)를 방지하기 위한 방법으로, 변수(차원)가 늘어날수록 과적합(Overfitting)의 위험성과 상관관계가 높은 변수로 인한 다중공선성 문제 등 예측이 불안정해지는 문제가 발생한다.
- PCA는 여러 개의 독립변수들을 잘 설명해줄 수 있는 주된 성분을 추출하는 방법으로, 이를 통해 핵심 특성만 선별하기 때문에 독립변수의 수(차원)를 줄일 수 있다.
- 주성분 분석을 위해서는 사용되는 변수들이 모두 양적변수여야 하고, 관측치들이 서로 독립적이고 정규분포를 이루고 있어야 한다.
- 변수의 수를 축약하면서 정보의 손실을 최소화하고자 할 때 주로 사용된다.
- PCA는 데이터 공간에 위치하는 점들의 분산을 최대한 보존하는 축을 통해 차원을 축소하는 것이 핵심이다.

### 2. Clustering Algorithm Practice

K-Means Clustering

```python
# 데이터 생성
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler

X, y = make_blobs(n_samples=500, n_features=2, centers=5, random_state=14)

X = StandardScaler().fit_transform(X)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
from sklearn.cluster import KMeans
kmc = KMeans(n_clusters=5,
            init='random',
            n_init=10,
            max_iter=100,
            random_state=14)
kmc.fit(X)
labels = kmc.labels_
print(labels)
```

```python
import pandas as pd
kmc_columns = ['comp1', 'comp2']
X_kmc_df = pd.DataFrame(X, columns = kmc_columns)
X_kmc_df['target']=y
X_kmc_df['label']=labels
X_kmc_df.head(5)
```

```python
# 타겟 클래스 종류
print(set(X_kmc_df['target']))
print(set(X_kmc_df['label']))
```

```python
# k means clustering 결과
df = X_kmc_df
markers = ['o', 'x', '^', 's', '*']

for i, mark in enumerate(markers):
  df_i = df[df['label']==i]
  target_i = i
  X1 = df_i['comp1']
  X2 = df_i['comp2']
  plt.scatter(X1, X2, marker=mark, label=target_i)

  plt.xlabel('component1')
  plt.ylabel('component2')
  plt.legend()
  plt.show()
```

```python
# 실제 결과
df = X_kmc_df
markers=['o', 'x', '^', 's', '*']

for i, mark in enumerate(markers):
  df_i = df[df['target']==i]
  target_i = i
  X1 = df_i['comp1']
  X2 = df_i['comp2']
  plt.scatter(X1, X2, marker=mark, label=target_i)

  plt.xlabel('component1')
  plt.ylabel('component2')
  plt.legend()
  plt.show()
```

DBSCAN

```python
# 데이터 생성
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=500, noise=0.05, random_state=14)

X = StandardScaler().fit_transform(X)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:,1])
plt.show()
```

```python
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=0.3).fit(X)
labels = db.labels_
print(labels)
```

Gaussian Mixture Model

```python
# 데이터 생성
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=500,
                  n_features=2,
                  centers=2,
                  random_state=14)

X = StandardScaler().fit_transform(X)
```

```python
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
from sklearn.mixture import GaussianMixture
gm = GaussianMixture(n_components=2, random_state=14)
gm.fit(X)
labels = gm.predict(X)
print(labels)
```

```python
import pandas as pd
columns = ['comp1', 'comp2']
X_gm_df = pd.DataFrame(X, columns=columns)
X_gm_df['target'] = y
X_gm_df['label'] = labels
X_gm_df.head(5)
```

```python
# 타겟 클래스 종류
print(set(X_gm_df['target']))
print(set(X_gm_df['label']))
```

```python
# 가우시안 혼합 모형 결과
df = X_gm_df
markers=['o', 'x']

for i, mark in enumerate(markers):
  df_i = df[df['label']==i]
  target_i=i
  X1 = df_i['comp1']
  X2 = df_i['comp2']
  plt.scatter(X1, X2, marker = mark, label=target_i)

plt.xlabel('component1')
plt.ylabel('component2')
plt.legend()
plt.show()
```

```python
# 실제 타겟 플랏
df = X_gm_df
markers=['o', 'x']

for i, mark in enumerate(markers):
  df_i = df[df['target']==i]
  target_i=i
  X1 = df_i['comp1']
  X2 = df_i['comp2']
  plt.scatter(X1, X2, marker = mark, label=target_i)

plt.xlabel('component1')
plt.ylabel('component2')
plt.legend()
plt.show()
```
